```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_csv('/content/sample_data/california_housing_test.csv')
dataset.head()
```

```python
colormap = np.array(['blue', 'orange', 'green','yellow','black'])
```

```python
def kmeancluster(k,parameters):
  if(parameters==0):
    x = dataset.iloc[:, [2, 8]].values
    plt.xlabel('housing_median_age')  
    plt.ylabel('median_house_value')
  elif(parameters==1):
    x = dataset.iloc[:, [5, 8]].values 
    plt.xlabel('population')  
    plt.ylabel('median_house_value')
  elif(parameters==2):
    x = dataset.iloc[:, [7, 8]].values 
    plt.xlabel('median_income')  
    plt.ylabel('median_house_value')
  model = KMeans(n_clusters=k)  
  y_predict= model.fit_predict(x)

  plt.title('K Mean Classification of housing') 
  for i in range(0,k):
    plt.scatter(x[y_predict == i, 0], x[y_predict == i, 1], s = 40, c = colormap[i]) 
  plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s = 100, c = 'red')   
  plt.show()
```

```python
kmeancluster(3,1) #median_house_value vs population, k=3
```

```python
kmeancluster(2,2) #median_house_value vs median_income, k=2
```

```python
kmeancluster(5,0) #median_house_value vs population , k=5
```
